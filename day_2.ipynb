```python
safe = 0

def is_increasing(lst):
    for i in range(0, len(lst)-1):
        if lst[i] < lst[i+1] and 1 <= abs(lst[i+1]-lst[i]) <= 3:
            continue
        else:
            return False
    return True

def is_decreasing(lst):
    for i in range(0, len(lst)-1):
        if lst[i] > lst[i+1] and 1 <= abs(lst[i+1]-lst[i]) <= 3:
            continue
        else:
            return False
    return True

for line in open('day_2_input.txt', 'r'):
    lst = list(map(int, line.split(' ')))
    if is_increasing(lst) or is_decreasing(lst):
        safe += 1

safe
```

# Day 2 Part 2

```python
safe = 0

def is_increasing(lst):
    for i in range(0, len(lst)-1):
        if lst[i] < lst[i+1] and 1 <= abs(lst[i+1]-lst[i]) <= 3:
            continue
        else:
            return False
    return True

def is_decreasing(lst):
    for i in range(0, len(lst)-1):
        if lst[i] > lst[i+1] and 1 <= abs(lst[i+1]-lst[i]) <= 3:
            continue
        else:
            return False
    return True

for line in open('day_2_input.txt', 'r'):
    lst = list(map(int, line.split(' ')))
    if is_increasing(lst) or is_decreasing(lst):
        safe += 1
    else:
        flag = False
        for i in range(len(lst)):
            new_lst = lst[:i] + lst[i+1:]
            if is_increasing(new_lst) or is_decreasing(new_lst):
                flag = True
                break
        if flag:
            safe += 1

safe
```

```python
from itertools import permutations

x = 4
valid = set()

for num_ps in range(0, x+1):
    sentence = 'p' * num_ps + 'n' * x + 'q'

    for perm in permutations(sentence):
        no_p = ''.join(perm).replace('p', '')
        N = len(no_p)
        if N % 2 == 0 or no_p.index('q') != len(no_p)//2:
            continue
        if no_p[:len(no_p)//2].count('n') != no_p[len(no_p)//2+1:].count('n'):
            continue
        valid.add(''.join(perm))

valid, len(valid)
```
