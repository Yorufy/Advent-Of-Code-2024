# Day 12 Part 1

The contribution of a square to its region's perimeter is just the number of adjacent squares which are not in the region. We will pad the grid with '0' so that the perimeter on the edge of the grid is accounted for.

```python
grid = []

for line in open('day_12_input.txt', 'r'):
    line = '0' + line.strip() + '0'
    grid.append(line)

grid.insert(0, '0'*len(line))
grid.append('0'*len(line))

ROWS = len(grid)
COLS = len(grid[0])


DIRECTIONS = [(0,1),(0,-1),(1,0),(-1,0)]
visited = [[False] * COLS for _ in range(ROWS)]
total_price = 0

def bfs(row: int, col: int) -> int:
    # Given a row and a column, return the price of fencing the region.

    q = [(row, col)]
    perimeter = 0
    area = 0

    while q:
        
        r, c = q.pop(0)
        if visited[r][c]:
            continue
        else:
            for dr, dc in DIRECTIONS:
                if 0 <= r+dr < ROWS and 0 <= c+dc < COLS:
                    if grid[r+dr][c+dc] == grid[r][c]:
                        q.append((r+dr, c+dc))
                    else: #Different character, so we can add to perimeter calculation
                        perimeter += 1

            visited[r][c] = True
            area += 1

    price = area * perimeter
    return price

for i in range(ROWS):
    for j in range(COLS):
        if visited[i][j] or grid[i][j] == '0':
            continue
        res = bfs(i, j)
        total_price += res

total_price
```

```python
grid = []

for line in open('day_12_input.txt', 'r'):
    line = '0' + line.strip() + '0'
    grid.append(line)

grid.insert(0, '0'*len(line))
grid.append('0'*len(line))

ROWS = len(grid)
COLS = len(grid[0])

DIRECTIONS = [(0,1),(0,-1),(1,0),(-1,0)]
visited = [[False] * COLS for _ in range(ROWS)]
total_price = 0

def bfs(row: int, col: int) -> int:
    # Given a row and a column, return the price of fencing the region.

    q = [(row, col)]
    REGION_CHAR = grid[row][col]
    area = 0
    region = set()

    while q:
        
        r, c = q.pop(0)
        if visited[r][c]:
            continue
        else:
            for dr, dc in DIRECTIONS:
                if 0 <= r+dr < ROWS and 0 <= c+dc < COLS:
                    if grid[r+dr][c+dc] == REGION_CHAR:
                        q.append((r+dr, c+dc))

            visited[r][c] = True
            region.add((r,c))
            area += 1

    sides = 0
    edges = dict()

    for r, c in region:
        for dr, dc in DIRECTIONS:
            nr, nc = r + dr, c + dc
            if (nr,nc) not in region:
                edges[((r+nr)/2,(c+nc)/2)] = (dr, dc)
    
    seen = set()
    for edge, direction in edges.items():
        if edge in seen:
            continue
        seen.add(edge)
        sides += 1
        er, ec = edge
        if er % 1 == 0: # Vertical edge
            for dr in [-1, 1]:
                i = 1
                while edges.get((er + i*dr, ec)) is not None and edges[(er+i*dr, ec)] == direction:
                    seen.add((er+i*dr, ec))
                    i += 1
        else: # Horizontal edge
            for dc in [-1, 1]:
                i = 1
                while edges.get((er, ec+i*dc)) is not None and edges[(er, ec+i*dc)] == direction:
                    seen.add((er, ec+i*dc))
                    i += 1

    price = area * sides
    return price


for i in range(ROWS):
    for j in range(COLS):
        if visited[i][j] or grid[i][j] == '0':
            continue
        res = bfs(i, j)
        total_price += res

total_price
```
